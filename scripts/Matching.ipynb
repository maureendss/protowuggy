```python
@staticmethod
def euclidean_distance(p, q):
    size = len(p) if hasattr(p, '__len__') else 1
    return np.linalg.norm(p-q) / size

def freq_loss(self):
    loss = self.euclidean_distance(self.EN.get_nb_spkr(), self.FR.get_nb_spkr())
    return loss/self.nb_spkr_norm
```
